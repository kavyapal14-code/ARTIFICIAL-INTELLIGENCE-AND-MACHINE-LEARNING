```python
 import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN

# Load Iris dataset from sklearn
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['species'] = data.target

# Show the first few rows
df.head()

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.drop('species', axis=1))

# Optional: Show scaled data
pd.DataFrame(X_scaled, columns=data.feature_names).head()

# K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Add labels to the dataframe
df['KMeans_Cluster'] = kmeans_labels

# Plot K-Means clustering
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'],
                hue=df['KMeans_Cluster'], palette='Set1')
plt.title('K-Means Clustering Results')
plt.show()

# Evaluate with silhouette score or metrics
from sklearn.metrics import silhouette_score
print("K-Means Silhouette Score:", silhouette_score(X_scaled, kmeans_labels))

# Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=3)
hierarchical_labels = hierarchical.fit_predict(X_scaled)

# Add labels to the dataframe
df['Hierarchical_Cluster'] = hierarchical_labels

# Plot Hierarchical clustering
plt.figure(figsize=(8, 6))
```
