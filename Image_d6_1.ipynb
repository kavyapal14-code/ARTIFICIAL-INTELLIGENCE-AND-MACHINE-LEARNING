```python
import cv2
import numpy as np
import matplotlib.pyplot as plt



```

```python
# Step 1: Load images
# cv2.imread(path) -> reads image in BGR format
original = cv2.imread('baboon.jpeg')
watermark = cv2.imread('image2.jpg')

# Step 2: Resize watermark to match original image size
# cv2.resize(image, (width, height))
watermark = cv2.resize(watermark, (original.shape[1], original.shape[0]))


# Step 3: Convert BGR to RGB for matplotlib display
# cv2.cvtColor(image, conversion_code)
original_rgb = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
watermark_rgb = cv2.cvtColor(watermark, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(original_rgb)
plt.title('Original Image')
plt.axis('off')
```

```python
plt.imshow(watermark_rgb)
plt.title('Watermark Image')
plt.axis('off')
```

```python
# Step 4: Apply watermark using alpha blending

# cv2.addWeighted(src1, alpha, src2, beta, gamma)
# Formula: result = src1 * alpha + src2 * beta + gamma

alpha = 0.3   # transparency level of watermark

watermarked = cv2.addWeighted(original, 1, watermark, alpha, 0)

# Convert for display
watermarked_rgb = cv2.cvtColor(watermarked, cv2.COLOR_BGR2RGB)
plt.imshow(watermarked_rgb)
plt.title('Watermarked Image')
plt.axis('off')
```

```python
# Step 5: Save watermarked image
# cv2.imwrite(filename, image)

cv2.imwrite('watermarked_image.jpg', watermarked)
```

```python
# Step 6: Compare original and watermarked images

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(original_rgb)
plt.title('Original')

plt.subplot(1, 2, 2)
plt.imshow(watermarked_rgb)
plt.title('Watermarked')

plt.show()
```
