```python
import cv2
import numpy as np
import matplotlib.pyplot as plt



```

```python
#read and display image
image= cv2.imread('baboon.jpeg')
print(image)
image.shape
```

```python
#convert to gray scale
gray = cv2.cvtColor (image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
gray.shape
```

```python
for m in range(0, 344):
    for n in range(0, 512):
        if gray[m, n] > 127:
            gray[m, n] = 255
        else:
            gray[m, n] = 0

print(gray)
```

```python
bw = np.zeros((512, 512))

for m in range(1, 512):
    for n in range(1, 512):
        if gray[m, n] >= 127:
            bw[m, n] = 255
        else:
            bw[m, n] = 0

print(bw)
plt.imshow(bw)
```
