```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Load original and watermarked image

original = cv2.imread('baboon.jpeg')
watermarked = cv2.imread('image2.jpg')

# Resize original to match watermarked image
original = cv2.resize(original, (watermarked.shape[1], watermarked.shape[0]))

# Convert to RGB for display
original_rgb = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
watermarked_rgb = cv2.cvtColor(watermarked, cv2.COLOR_BGR2RGB)
```

```python
alpha = 0.3  # same alpha used earlier

# cv2.subtract(src1, src2) -> subtract pixel values
extracted = cv2.subtract(watermarked, original)

# Divide by alpha to recover watermark
extracted = extracted / alpha

# Clip values to valid range (0-255)
# np.clip(array, min, max)
extracted = np.clip(extracted, 0, 255).astype(np.uint8)
```

```python
extracted = np.clip(extracted, 0, 255).astype(np.uint8)

# Convert for display
extracted_rgb = cv2.cvtColor(extracted, cv2.COLOR_BGR2RGB)

plt.imshow(extracted_rgb)
plt.title('Extracted Watermark')
plt.axis('off')
```
